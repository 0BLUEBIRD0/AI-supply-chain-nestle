Section 1: Key Findings on Nestlé India's Supply Chain Challenges
Nestlé India operates a vast supply chain handling over 4,000 FMCG products, including staples like Maggi noodles, Nescafé coffee, KitKat chocolates, Milkmaid condensed milk, and Cerelac infant nutrition. The Indian market presents unique challenges due to its scale (1.4B population, 70% rural), regulatory environment (e.g., GST, FSSAI compliance), and external factors like climate variability. Based on recent analyses (Nestlé Annual Report 2024-25, industry reports), here are the core challenges:
Summary Table of Challenges:
| Challenge Category       | Specific to Nestlé India FMCG                          | Potential Impact                          |
|--------------------------|-------------------------------------------------------|-------------------------------------------|
| Logistics/Distribution  | Rural-urban fragmentation (500+ distributors), monsoon delays, quick commerce demands (e.g., 10-15 min deliveries via Blinkit) | 15-20% higher logistics costs; replenishment delays leading to stockouts |
| Demand Forecasting      | Seasonal surges (e.g., 20-30% Diwali boost for Maggi), perishability of dairy/chocolates | 10-15% stockouts; ₹500-1,000 Cr annual waste from overstock |
| Inventory Management    | High holding costs for perishables, balancing urban e-com vs. rural kirana stores | 14-20% excess inventory; sustainability risks (e.g., food waste) |
| Sustainability & Sourcing | Deforestation-free cocoa/palm oil targets by 2025, water scarcity in sourcing regions | Sourcing delays; potential ESG fines/compliance issues |

Key Insights:

Nestlé is investing ₹5,000 crore in supply chain expansions by 2025, focusing on digital twins and AI for agility (e.g., new warehouses in Tier 2 cities).
Post-COVID, logistics costs rose 15-20% globally, with India facing amplified issues from fuel volatility and e-commerce growth (quick commerce projected at $5.4B by end-2025).
AI Opportunities: Nestlé's global NesGPT (LLM for predictions) and SAS partnerships are piloting in India for automated forecasting, reducing errors by 30% and safety stock by 14-20%.

These findings highlight levers for AI intervention: accurate forecasting to tame variability, optimization models for inventory/routing, and simulations for resilience.
Section 2: Project Objectives & Scope
SMART Objectives

Demand Forecasting: Build an LSTM-based AI model to predict monthly sales for top 5 products (Maggi, Nescafé, KitKat, Milkmaid, Cerelac) with <10% MAPE, incorporating Indian factors like monsoons and festivals.
Inventory Optimization: Develop a linear programming model (using PuLP) to recommend stock levels, targeting 15-20% reduction in holding costs while achieving 95% service levels.
Routing Efficiency: Implement route optimization (using OR-Tools) for key hubs (e.g., Delhi-Mumbai), aiming for 10-15% logistics cost savings in quick commerce scenarios.
Business Impact Quantification: Simulate end-to-end optimizations to project annual savings (e.g., ₹X Cr) and sustainability gains (e.g., 5% waste reduction), with sensitivity to disruptions.

Scope

In Focus: North India simulation (3-5 hubs, 2024-25 data period); core processes (forecasting, inventory, routing).
Assumptions: Use simulated/synthetic data mimicking real Nestlé patterns (e.g., Kaggle FMCG datasets adapted).
Out of Scope: Real-time implementation, full end-to-end ERP integration, hardware/IoT sensors.

Success Metrics

Forecasting: MAPE <10%, MAE.
Optimization: Cost savings %, Inventory Turnover Ratio >4x, Service Level >95%.
Overall: Simulated ROI (e.g., payback <6 months).
